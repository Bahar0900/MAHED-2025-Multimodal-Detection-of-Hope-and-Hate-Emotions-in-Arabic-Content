```python
import emoji
import re
from googletrans import Translator

def extract_emojis(text):
    emoji_pattern = re.compile("[\U00010000-\U0010ffff]", flags=re.UNICODE)
    return emoji_pattern.findall(text)

def replace_emojis_with_arabic(text):
    translator = Translator()
    emojis_found = extract_emojis(text)
    for emj in set(emojis_found):  # use set to avoid repeating translations
        desc = emoji.demojize(emj).replace(":", "").replace("_", " ")
        try:
            arabic_desc = translator.translate(desc, dest='ar').text
        except Exception:
            arabic_desc = desc  # fallback if translation fails
        text = text.replace(emj, f" {arabic_desc} ")
    return text



# Process
text = replace_emojis_with_arabic(text)
print(text)
```

```python

import re
import emoji
from googletrans import Translator
from camel_tools.utils.normalize import normalize_unicode, normalize_alef_maksura_ar, normalize_teh_marbuta_ar
from pyarabic import araby
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Ensure nltk resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import re
import emoji
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from pyarabic import araby
from camel_tools.utils.normalize import normalize_unicode, normalize_alef_maksura_ar, normalize_teh_marbuta_ar

# Preload stopwords once
arabic_stopwords = set(stopwords.words('arabic'))

def clean_arabic_text(text):
    # Remove RTs and mentions
    text = re.sub(r"RT\s?@\w+:", "", text)
    text = re.sub(r"@\w+", "", text)
    
    # Remove URLs
    text = re.sub(r"http\S+|www.\S+", "", text)
    
    # Remove all emojis (no translation)
    text = emoji.replace_emoji(text, replace="")
    
    # Normalize Arabic text
    text = normalize_unicode(text)
    text = normalize_alef_maksura_ar(text)
    text = normalize_teh_marbuta_ar(text)
    
    # Remove diacritics (Tashkeel)
    text = araby.strip_tashkeel(text)
    
    # Keep only Arabic characters and spaces
    text = re.sub(r"[^\u0600-\u06FF\s]", "", text)
    
    # Tokenize and remove stopwords
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in arabic_stopwords]
    
    return " ".join(tokens)
```

```python
text = "'@thaer_alnashef سؤال اخ ثائر لماذا يتم التطبيع فى هذه الظروف بالذات والوطن العربى ممزق هل هذه جنه المسيخ الدجال ام اسرائيل هى من تسببت فى فوضى المنطقه والتطبيع هو الثمن لإعادتها مرة اخرى كما كانت'"
text = clean_arabic_text(text)
print(text)
```
